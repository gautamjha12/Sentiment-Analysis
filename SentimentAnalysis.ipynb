```python
import pandas as pd
data =  pd.read_csv('imdb.csv')
df = pd.DataFrame(data)
pd.set_option('display.max_colwidth',None)
df = df.apply(lambda x: x.astype(str).str.lower())
df["review"] = df['review'].str.replace('[^\w\s\w*\d]','')
df["review"] = df['review'].str.replace('[0-9]','')
df
```

```python
import re
def tokenizer(df):
    token = re.split('\W+',df)
    return token

df['tokenized_reviews'] = df['review'].apply(lambda x :tokenizer(x))    
```

```python
df.head()
```

```python
import nltk
stopwords = nltk.corpus.stopwords.words('english')
def remove_stopwords(tokenized_reviews):
    cleaned_text = [i for i in tokenized_reviews if i not in stopwords]
    return cleaned_text

df['rem_stopword'] = df['tokenized_reviews'].apply(lambda x : remove_stopwords(x))
```

```python
df.head()
```

```python
import nltk
from nltk.stem import PorterStemmer
ps = PorterStemmer()

#Stemming

def stemming(tokenized_reviews):
    z = [ps.stem(i) for i in tokenized_reviews]
    return z
 
df['stemmed_words'] = df['rem_stopword'].apply(lambda x : stemming(x))
```

```python
df.head()
```

```python
wn = nltk.WordNetLemmatizer()

def lemmatizer(tokenized_reviews):
    lemmatize = [wn.lemmatize(word) for word in tokenized_reviews]
    return lemmatize

df['lemmatize'] = df['rem_stopword'].apply(lambda x: lemmatizer(x))
```
